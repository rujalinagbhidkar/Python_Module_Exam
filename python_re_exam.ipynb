```python
#Find the Second Largest Number in the list 
num = [10,5,20,8]
largest = second = 0
for n in num:
    if n > largest:
        second = largest
        largest = n
    elif n > second and n != largest:
        second = n
print(second)
```

```python
#Find the Largest Word in the Sentence
sentence = "Big data analytics is powerful"
words = sentence.split()
#print(words)
long = words[0]
#print(long)
for i in words:
    if len(i) > len(long):
        long = i

print(long)
```

```python
#Find Student with the Highest marks in the dictionary 
students = {'Amit': 78, 'Neha': 85, 'Ravi': 82}
high = max(students, key = students.get)
print(high)
```
