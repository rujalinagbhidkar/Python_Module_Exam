```python
#QUESTION 1
def longest_increasing_subsquence(a):
    inc_seq = []
    inc_seq.append(min(a))
    inc = 1
    ele = 0
    for i in inc_seq:
        ele += i + (inc**2)
        if ele in a:
            inc_seq.append(ele)
        
    count = len(inc_seq)
    print(f"Count_lis : {count}")
    return inc_seq 

arr = [10,9,2,5,3,7,101,18]
longest_increasing_subsquence(arr)
```

```python
# Questio _2

def fibonacci_sequence(t1, t2, n, k):
    t1,t2 = t1,t2
    list = []
    list.append(t1)
    list.append(t2)
    for i in range(1,n-1):
        t3 = t2 + t1
        list.append(t3)
        t1 = t2
        t2 = t3
    fm = list[-1]
    print(f"fibonacci_modified: {fm}")

    sum = 0
    for i in list:
        sum += i
    print(f"Sum_range(1,5): {sum}")

    count = 0
    for i in range(1,n-1):
        if i % k == 0:
            count += 1
    print(f"Count_divisible : {count}")
    return 

fibonacci_sequence(0,1,5,2)

                  
```
